2. Using Line Graph finding no. of employees working as part time & Full Time in

```python
part_time_emp = wages[(wages.type == 'Part Time')]
part_time_emp1 = part_time_emp.groupby('industry')['type'].count()
part_time_emp1

full_time_emp = wages[(wages.type == 'Full Time')]
full_time_emp1 = full_time_emp.groupby('industry')['type'].count()
full_time_emp1

print("\n\nThe Part Time Male Employees:",part_time_emp1)
print("\n\nThe Full Time Male Employees:",full_time_emp1)

f,ax = plt.subplots(figsize=(10, 8))
line_chart1 = plt.plot(range(1,13), full_time_emp1)
line_chart2 = plt.plot(range(1,13), part_time_emp1)
plt.title('Line Graph for Full Time Vs Part Time Employees',fontsize=18 )
plt.xlabel('Different Industries')
plt.ylabel('No. of Employees')
plt.legend(['Full Time', 'Part Time'], loc=2)
plt.show()
from IPython.core.display import HTML as Center

Center(""" <style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style> """)

```

```python
part_time_fem = wages[(wages.type == 'Part Time') & (wages.sex == 'female')]
part_time_fem1 = part_time_fem.groupby('industry')['type'].count()
part_time_fem1

len(part_time_fem1)

full_time_fem = wages[(wages.type == 'Full Time') & (wages.sex == 'female')]
full_time_fem1 = full_time_fem.groupby('industry')['type'].count()
full_time_fem1
len(full_time_fem1)

part_time_male = wages[(wages.type == 'Part Time') & (wages.sex == 'male')]
part_time_male1 = part_time_male.groupby('industry')['type'].count()
part_time_male1

full_time_male = wages[(wages.type == 'Full Time') & (wages.sex == 'male')]
full_time_male1 = full_time_male.groupby('industry')['type'].count()
full_time_male1

print("The Part Time Female Employees:",part_time_fem1)
print("\n\nThe Full Time Female Employees:",full_time_fem1)
print("\n\nThe Part Time Male Employees:",part_time_male1)
print("\n\nThe Full Time Male Employees:",full_time_male1)

f,ax = plt.subplots(figsize=(10, 8))
line_chart1 = plt.plot(range(1,13), full_time_male1,'g')
line_chart2 = plt.plot(range(1,13), full_time_fem1,'b')
line_chart3 = plt.plot(range(1,13), part_time_male1,'r')
line_chart4 = plt.plot(range(1,12), part_time_fem1,'m')
plt.title('Full & Part Time, Male and Female Employees',fontsize=18 )
plt.xlabel('Different Industries')
plt.ylabel('No. of Employees')
plt.legend(['FT Males','FT Females','Part Time Males','Part Time Females'], loc=2)
plt.show()
```

****************************************************************************
4. Barchart - For Full Time & Part Time Employment by gender.
*******************************************************************************

```python
tot_part_full = wages[['sex','type', "industry"]].groupby(['sex','type']).count()
tot_part_full

print("\n\nThe Total Part - Full Time Male Employees:",tot_part_full)

emp_pf=['Female_FT','Female_PT','Male_FT','Male_PT']
pos=np.arange(len(tot_part_full['industry']))

count = tot_part_full['industry']
f,ax = plt.subplots(figsize=(10, 8))
plt.bar(pos,count,color='blue',edgecolor='black')
plt.xticks(pos, emp_pf)

plt.xlabel('Gender Wise', fontsize=16)
plt.ylabel('No. of Employees', fontsize=16)
plt.title('Barchart - Full Time & Part Time Employees by Gender',fontsize=16)
plt.show()
```

*******************************************************************************
10. Using Grouped Stacked bar finding Graduates, Under Graduate and 
Highschool or less among males and females.
********************************************************************************

```python
Grad_fem = wages[(wages.sex == 'female') & (wages.education >= 17)]
Grad_fem1 = len(Grad_fem['education'])

under_Grad_fem = wages[(wages.sex == 'female') & (wages.education >=16)& (wages.education < 17)]
under_Grad_fem1 = len(under_Grad_fem['education'])

hs_fem = wages[(wages.sex == 'female') & (wages.education <= 15)]
hs_fem1 = len(hs_fem['education'])


Grad_male = wages[(wages.sex == 'male') & (wages.education >= 17)]
Grad_male1 = len(Grad_male['education'])

under_Grad_male = wages[(wages.sex == 'male') & (wages.education >=16)& (wages.education < 17)]
under_Grad_male1 = len(under_Grad_male['education'])

hs_male = wages[(wages.sex == 'male') & (wages.education <= 15)]
hs_male1 = len(hs_male['education'])

edu=['Graduate', 'Under Graduate', 'Secondary School or less']

Gender=['Male','Female']

pos1=np.arange(len(edu))
bar_width = 0.35

edu_male = [Grad_male1, under_Grad_male1, hs_male1]
edu_fem =  [Grad_fem1, under_Grad_fem1, hs_fem1]
f,ax = plt.subplots(figsize=(10, 8))
plt.bar(pos1,edu_male,bar_width,color='red',edgecolor='black')
plt.bar(pos1+bar_width,edu_fem,bar_width,color='blue',edgecolor='black')
plt.xticks(pos1, edu)
plt.xlabel('Education', fontsize=16)
plt.ylabel('Count of  Males & Females', fontsize=16)
plt.title('Males & Females Education',fontsize=18)
plt.legend(Gender,loc=2)

plt.show()
```
